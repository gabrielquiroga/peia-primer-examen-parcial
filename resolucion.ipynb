### Ejercicio 1
(2 puntos) De 10 monedas hay 6 monedas falsas, que tienen probabilidad 0,5 de mostrar cara al ser lanzadas. Si se toma una moneda al azar, se lanza 12 veces, y en todas ellas se obtiene cara, ¿qué es más probable, que la moneda elegida sea justa o que esté cargada? Dar las probabilidades tanto de que la moneda elegida sea falsa, como de que sea justa.

Definamos los eventos, tales que: 
- F -> Moneda falsa
- J -> Moneda justa
- C -> Obtener cara en un lanzamiento
- E -> Obtener 12 caras consecutivas

Además, como la probabilidad de que una moneda falsa muestre cara al ser lanzada es igual a la de una moneda justa en este caso (0,5), mejor representaremos dicha probabilidad como *x*, para plantear el problema de manera genérica. Entonces:
- $P(F) = 0,6$
- $P(J) = 0,4$
- $P(C|F) = x$
- $P(C|J) = 0,5$

Busquemos la probabilidad, para cada tipo de moneda, de obtener cara en 12 lanzamientos. Primero, para la moneda falsa:
- Por teorema de Bayes -> $P(F|E) = \frac{P(E|F) P(F)}{P(E)}$

donde $P(E)$ se obtiene a partir de la ley de probabilidad total:
$$P(E) = P(E|F)P(F) + P(E|J)P(J)$$

E igualmente para una moneda justa:
- $P(J|E) = \frac{P(E|J) P(J)}{P(E)}$ ; donde $P(E)$ puede se obtiene por la fórmula anterior

Por otro lado, tenemos las siguientes probabilidades condicionales, dado que cada lanzamiento es independiente y la probabilidad no cambia entre lanzamientos:
- $P(E|F) = P(C|F)^{12} = x^{12}$
- $P(E|J) = P(C|J)^{12} = 0.5^{12} = 0.000244140625$

Resolvemos entonces:
- $P(E) = P(E|F)P(F) + P(E|J)P(J) = x^{12}\cdot 0.6 + 0.000244140625\cdot 0.4 = 0.00009765625 + 0.6\cdot x^{12}$

- $P(F|E) = \frac{x^{12}\cdot 0.6}{0.00009765625 + 0.6\cdot x^{12}}$

- $P(J|E) = \frac{0.000244140625\cdot 0.4}{0.00009765625 + 0.6\cdot x^{12}}$

Dándole el valor 0,5 a *x*:
- $P(F|E) = \frac{0.000244140625\cdot 0.6}{0.00009765625 + 0.6\cdot 0.000244140625} = \frac{0.000146484375}{0.000244140625} = 0.6$

- $P(J|E) = \frac{0.000244140625\cdot 0.4}{0.00009765625 + 0.6\cdot x^{12}} = 0.4$

Lo cual tiene sentido, si las monedas falsas y justas tienen la misma probabilidad de mostrar cara.




```python
# Calculadora de probabilidades para distinto x:
x = 0.5 # Elegir la probabilidad para la moneda falsa de salir cara (entre 0 y 1)
P_E = 0.00009765625 + 0.6 * x**12
P_FE = (0.6 * x**12)/P_E
P_JE = (0.4 * 0.000244140625)/P_E

print("Probabilidad de que la moneda sea falsa dado que salió 12 veces cara:", P_FE)
print("Probabilidad de que la moneda sea justa dado que salió 12 veces cara:", P_JE)
```

#### a) Determine el valor de K
Buscamos los valores de $x$ para los cuales $f_{X,Y}$ es distinta de 0, para lo cual calculamos los valores que satisfacen $100x^2 = 8x$. A simple vista, se puede obtener: $x_1 = 0$ y $x_2 = \frac{8}{100} = 0.08$.

Podemos asegurar ahora que $x \in [0 ; 0.08]$ y que $y \in [100x^2 ; 8x]$. Luego, planteamos la integral sobre toda la región para la función de densidad conjunta, que debe dar como resultado 1.
$$
\int_0^{0.08}\int_{100x^2}^{8x} K y dy dx = 1
$$
Operando:
$$
\int_0^{0.08}\int_{100x^2}^{8x} K y dy dx = K\int_0^{0.08} (\frac{y^2}{2})|_{100x^2}^{8x} = K\int_0^{0.08} (32x^2 - 5000x^4)dx = K(\frac{32}{3}x^3 - \frac{5000}{5}x^5)|_0^{0.08} =
$$
$$
= K(\frac{32}{3}0.08^3 - \frac{1000}{1} 0.08^5) = K\frac{512}{234375} = 1
$$
Despejando $K$:
$$
K = \frac{234375}{512} \approx 457.764
$$

#### b) Encuentre la densidad marginal $f_X(x)$ de $X$
Para encontrar la densidad marginal de $f_X(x)$ de X, partimos de:
$$
f_X(x) = \int_{c(x)}^{d(x)} f_{X, Y}(x, y) dy
$$
Reemplazando por la función e intervalo correspondientes:
$$
f_X(x) = \int_{100x^2}^{8x} Ky dy = K (\frac{y^2}{2})|_{100x^2}^{8x} = K (32x^2 - 5000x^4) = \frac{234375}{512}(32x^2 - 5000x^4)
$$
Esta función sólo será válida para el intervalo de $x$ indicado y valdrá $0$ en otro caso. Entonces:
$$
f_X(x) = \begin{cases}
\frac{234375}{512}(32x^2 - 5000x^4) &; x \in [0; 0.08] \\
0 &; \text{en otro caso}
\end{cases}
$$

#### c) Encuentre la densidad marginal $f_Y(y)$ de $Y$
Similarmente al punto anterior, partimos de:
$$
f_Y(y) = \int_{e(y)}^{f(y)} f_{X, Y}(x, y) dx
$$
pero en este caso aún tenemos que definir los intervalos. Evaluamos la función $8x$ para $x = 0$ y $x = 0.08$, así obtenemos que $y$ debe estar en el intervalo $[0; 0.64]$.
Así también podemos inferir desde $100x^2 \leq y \leq 8x$ el intervalo de x:
$$
x \in [\frac{y}{8} ; \frac{\sqrt{y}}{10}]
$$
Operando de la misma manera:
$$
f_Y(y) = \int_{\frac{y}{8}}^{\frac{\sqrt{y}}{10}} K y dx = K y (x)|_{\frac{y}{8}}^{\frac{\sqrt{y}}{10}} = K y (\frac{\sqrt{y}}{10} - \frac{y}{8}) = \frac{234375}{512}y (\frac{\sqrt{y}}{10} - \frac{y}{8})
$$
Teniendo en cuenta el intervalo de validez resulta:
$$
f_Y(y) = \begin{cases}
\frac{234375}{512}y (\frac{\sqrt{y}}{10} - \frac{y}{8}) &; y \in [0; 0.64] \\
0 &; \text{en otro caso}
\end{cases}
$$

#### d) Encuentre la densidad condicional $f_{X|Y}(x|y)$ de $X$ dado $Y$
Por definición, tenemos que la densidad condicional es:
$$
f_{X|Y}(x|y) = \frac{f_{X, Y}(x, y)}{f_Y(x)} 
$$
donde ya se conoce de antes:
$$f_{X, Y}(x, y) = Ky = \frac{234375}{512}y$$
$$f_Y(x) = Ky (\frac{\sqrt{y}}{10} - \frac{y}{8}) = \frac{234375}{512}y (\frac{\sqrt{y}}{10} - \frac{y}{8})$$
Reemplazando:
$$f_{X|Y}(x|y) = \frac{Ky}{Ky (\frac{\sqrt{y}}{10} - \frac{y}{8})} = \frac{1}{(\frac{\sqrt{y}}{10} - \frac{y}{8})}$$
Sin olvidar los intervalos válidos de $x$ e $y$, mencionados en **c)**

#### e) Encuentre la densidad condicional $f_{Y|X}(y|x)$ de $Y$ dado $X$
Nuevamente, por definición:
$$
f_{Y|X}(y|x) = \frac{f_{X, Y}(x, y)}{f_X(x)} 
$$
donde ya se conoce de antes:
$$f_{X, Y}(x, y) = Ky = \frac{234375}{512}y$$
$$f_X(x) = K (32x^2 - 5000x^4) = \frac{234375}{512}(32x^2 - 5000x^4)$$
Reemplazando:

### Ejercicio 3
(2 puntos) Sea:
$$Y = R\cos(2\pi·\Phi)$$
donde $R$ y $\Phi$ son variables aleatorias independientes. La fase $\Phi$ se distribuye uniforme sobre el intervalo $(0, 1)$, mientras que la amplitud $R$ tiene una distribución de Rayleigh con densidad $f(r) = re^{−r^2/2}$ para $r > 0$. Muestre que $Y$ tiene una distribución normal.

Pista: Defina $Y = R\cos(2\pi \cdot \Phi)$ y $X = R\sin(2\pi \cdot \Phi)$ y utilice la relación $X^2 + Y^2 = R^2$ para calcular la densidad conjunta de $(X, Y)$ a partir de la densidad conjunta de $(R, \Phi)$.

Se tienen las densidades marginales de las variables aleatorias $R$ y $\Phi$:
- $f_R(r) = re^{−r^2/2}$
- $f_\Phi(\phi) = \frac{1}{1-0} = 1$

Al ser las variables  $R$ y $\Phi$ independientes, su función de densidad conjunta es el producto de sus densidades marginales:
$$f_{R, \Phi}(r, \phi) = re^{−r^2/2}$$

Se define la transformación de variables $(R, \Phi) \rightarrow (X, Y)$:
- $Y = R\cos(2\pi \cdot \Phi)$
- $X = R\sin(2\pi \cdot \Phi)$

tal que la función de transformación inversa, entonces será:
- $R = \sqrt{X^2 + Y^2}$
- $\Phi = \frac{1}{2\pi}\tan^{-1}{(\frac{x}{y})}$

La densidad conjunta de $(X, Y)$ se puede obtener a partir de la densidad conjunta de $(R, \Phi)$ según en cálculo:
$$f_{X,Y}(x,y) = f_{R,\Phi}(r,\phi)\cdot\left|J\right|$$
donde $\left|J\right|$ es el valor abosulto del determinante Jacobiano de la transformación, el cual se calcula como:
$$J = \begin{pmatrix} \frac{\partial r}{\partial x} & \frac{\partial r}{\partial y} \\ \frac{\partial \phi}{\partial x} & \frac{\partial \phi}{\partial y} \end{pmatrix}$$

$$\frac{\partial r}{\partial x} = \frac{\partial }{\partial x} \sqrt{x^2 + y^2} = \frac{x}{\sqrt{x^2 + y^2}}$$
$$\frac{\partial r}{\partial y} = \frac{\partial }{\partial y} \sqrt{x^2 + y^2} = \frac{y}{\sqrt{x^2 + y^2}}$$
$$\frac{\partial \phi}{\partial x} = \frac{\partial }{\partial x}\frac{1}{2\pi}\tan^{-1}\left(\frac{x}{y}\right) = \frac{1}{2\pi} \frac{\frac{x}{y}}{1 + \left(\frac{x}{y}\right)^2} = \frac{1}{2\pi} \frac{y^2}{x^2 + y^2} \frac{1}{y} = \frac{y}{2\pi \left(x^2 + y^2\right)}$$
$$\frac{\partial \phi}{\partial y} = -\frac{x}{2\pi \left(x^2 + y^2\right)}$$

Entonces el determinante se calcula como:
$$\det (J) = \frac{\partial r}{\partial x} \frac{\partial \phi}{\partial y} - \frac{\partial r}{\partial y} \frac{\partial \phi}{\partial x} = \frac{x}{\sqrt{x^2 + y^2}} \left(-\frac{x}{2\pi \left(x^2 + y^2\right)}\right) - \frac{y}{\sqrt{x^2 + y^2}} \frac{y}{2\pi \left(x^2 + y^2\right)}$$
$$ = - \left(\frac{x^2 + y^2}{2\pi \sqrt{x^2 + y^2}\left( x^2 + y^2\right)}\right) = -\frac{1}{2\pi \sqrt{x^2 + y^2}}$$

Que en valor absoluto es:
$$\left| J\right| = \frac{1}{2\pi \sqrt{x^2 + y^2}}$$

A posteriori, se puede ver que hubiese resultado más fácil calcular el determinante Jacobiano de la transformación directa y luego simplemente invertirlo: $\left|J \right| = \frac{1}{\left|J^{-1} \right|}$

Ya se tiene lo necesario para calcular la desidad conjunta de $(X, Y)$:
$$f_{X,Y}(x,y) = f_{R,\Phi}(r,\phi)\cdot\left|J\right| = re^{\frac{−r^2}{2}}\frac{1}{2\pi \sqrt{x^2 + y^2}} = \sqrt{x^2 + y^2} e^{\frac{-\left(\sqrt{x^2 + y^2} \right)^2}{2}}\frac{1}{2\pi \sqrt{x^2 + y^2}}$$

Que simplificando términos y operando resulta:
$$f_{X,Y}(x,y) = \frac{1}{2\pi} e^{-\frac{\left(x^2 + y^2 \right)}{2}}$$


Ahora, con el objetivo de conocer la distribución de $Y$, buscamos la densidad marginal de $Y$. De la densidad conjunta obtenida $f_{X,Y}(x,y)$ podemos redistribuir algebraicamente:
$$f_{X,Y}(x,y) = \frac{1}{2\pi} e^{-\frac{\left(x^2 + y^2 \right)}{2}} = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}} \frac{1}{\sqrt{2\pi}} e^{-\frac{y^2}{2}}$$

en donde, dado que la densidad conjunta se factoriza como producto de una función exclusivamente dependiente de $x$ y otra función exclusivamente dependiente de $y$, podemos concluir $X$ e $Y$ independientes, y extraer las densidades marginales para cada variable como dichas funciones.
- $f_X(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}$
- $f_Y(y) = \frac{1}{\sqrt{2\pi}} e^{-\frac{y^2}{2}}$

La distribución normal, con parámetros $\mu , \sigma^2$, por definición es
$$\frac{1}{\sigma \sqrt{2\pi}} e^{\frac{-\left(x - \mu \right)^2}{2\sigma^2}}$$

Entonces, podemos afirmar que $Y$ tiene una distribución normal con parámetros varianza $\sigma^2 = 1$ y media $\mu = 0$.

```python
import numpy as np
import pandas as pd
import datetime
from scipy.stats import poisson
```

```python
# Consigna 1:
# Debería generar una tabla con todos los días de los 3 años mencionados (2023, 2024 y 2025), es decir 365 * 3 = 1095 registros (en caso sin bisiestos,
# esto es ilustrativo). El lambda para cada registro debe ser la suma de los efectos del año, del mes y del día de la semana correspondientes
# a la fecha del registro. Voy a crear un dataframe con la fecha completa y sus datos correspondientes (año, mes, día de la semana). Luego, debería crear
# 3 diccionarios para mapear los efectos y poder calcular el valor de lambda para cada registro, y agregarlo al dataframe.

start_date = datetime.datetime(2023, 1, 1)
end_date = datetime.datetime(2025, 12, 31)
index = pd.date_range(start_date, end_date)

df = pd.DataFrame({'date': index})
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['dayOfWeek'] = df['date'].dt.day_of_week
```

```python
year_effect_map = {
    2023: 1000,
    2024: 1500,
    2025: 2000
}
month_effect_map =  {
    1: 1000,
    2: 1500,
    3: 2000,
    4: 2000,
    5: 2500,
    6: 2500,
    7: 3000,
    8: 2500,
    9: 2500,
    10: 2000,
    11: 1500,
    12: 1000
}
# Usando pandas.DateTime el día 0 corresponde al Lunes y el 6 al Domingo
dof_effect_map = {
    6: 1000,
    0: 2000,
    1: 3000,
    2: 3500,
    3: 3000,
    4: 2000,
    5: 1000
}

df['year_effect'] = df['year'].map(year_effect_map)
df['month_effect'] = df['month'].map(month_effect_map)
df['dof_effect'] = df['dayOfWeek'].map(dof_effect_map)
```

```python
df['lambda'] = df['year_effect'] + df['month_effect'] + df['dof_effect']
```

```python
# Ahora tengo que generar ventas aleatorias para cada comerciante, siguiendo
# la distribución de Poisson que corresponde a cada día.
rng = np.random.default_rng(seed=42)

ventas_francisco = poisson.rvs(mu=df['lambda'], random_state=rng)
ventas_miguel = poisson.rvs(mu=df['lambda'], random_state=rng)

df['sales_francisco'] = ventas_francisco
df['sales_miguel'] = ventas_miguel
```

```python
# Consigna 2:
# Ahora se debe generar la función empírica de distribución y de densidad aproximada
# de las ventas por año (serían 3 gráficos), solo para Don Francisco.
# Se filtran por año y comerciante, se ordenan los valores y se obtienen las funciones
# con sus gráficos.

years = [2023, 2024, 2025]
colors = ['red', 'blue', 'green']
fig, axes = plt.subplots(3, 1, figsize=(8, 16))

for ax, year, color in zip(axes, years, colors):
    # Función empírica de distribución
    df_ventas = df[df['year'] == year]['sales_francisco'].sort_values()

    n = len(df_ventas)
    prob = np.arange(1, n + 1) / n

    ax.step(df_ventas, prob, where='post', color=color, label="Distribución acumulada")
    ax.set_title(f"Función empírica de distribución para ventas Don Francisco - {year}")
    ax.set_xlabel("Número de eventos")
    ax.set_ylabel("Probabilidad acumulada")
    ax.grid()
    ax.legend()
    ax.set_xlim(0, 10000)
    ax.set_ylim(0, 1.1)

    # Función de densidad aproximada (Histograma)
    ax_dens = ax.twinx()
    ax_dens.hist(df_ventas, bins=50, density=True, color=color, alpha=0.3, label="Densidad aproximada")
    ax_dens.set_ylabel("Densidad")
    ax_dens.set_ylim(0, 0.001)

    # Función de densidad aproximada por Kernel (KDE)
    sns.kdeplot(df_ventas, ax=ax_dens, color='black', fill=True, alpha=0.1, label='Densidad aproximada (KDE)')

    lines, labels = ax.get_legend_handles_labels()
    lines2, labels2 = ax_dens.get_legend_handles_labels()
    ax.legend(lines + lines2, labels + labels2, loc='upper left')

plt.show()
```
