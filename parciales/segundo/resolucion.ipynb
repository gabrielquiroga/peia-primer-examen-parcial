### Ejercicio 1
(3 puntos) Una variable aleatoria discreta Xpuede tomar los valores 0, 1, 2 y 3. Las probabilidades para cada valor posible
están dadas por la siguiente tabla:
|  X  |  0  |  1  |  2  |  3  |
| --- | --- | --- | --- | --- |
| p | $\frac{1\theta}{3}$ | $\frac{2\theta}{3}$ | $\frac{1 - 1\theta}{3}$ | $\frac{2\left(1 - 1\theta\right)}{3}$ |

Si experimentalmente se obtienen los siguientes datos:
(0,3,3,1,1,3,3,0,2,2),
determine el valor de $\theta$ usando el método de máxima verosimilitud.
